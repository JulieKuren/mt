## How often do we find a specific relation in a text?

```python
relation_common = dict.fromkeys(relation_classes, 0) # creating a dict, where all the values = 0

def check_relations(text):  
    words = nltk.word_tokenize(text)
    words_low = set([word.lower() for word in words]) # lowercase words from a text
    for key in relation_common.keys():
        if key in words_low:
            relation_common[key] += 1
    #return relation_common

for file in all_files:
    with open(os.path.join(directory, file), 'r') as f:
        text = f.read()
        check_relations(text)
        
for key in sorted(relation_common, key = relation_common.get, reverse = True):
    print(key, relation_common[key])
```
