```python
import os
import subprocess

from unidecode import unidecode
```

```python
class RACEParser:
    def __init__(self, input_dir, output_dir):
        self.input_dir = input_dir
        self.tmp_dir = "/tmp"
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        self.output_dir = output_dir
    
    @staticmethod
    def extract_article(src, dest):
        with open(src, "rt") as f_i:
            with open(dest, "wb") as f_o:
                f_o.write(unidecode(eval(f_i.read())["article"]).encode("ascii"))
        
    def parse_text(self, file_name):
        tmp_path = os.path.join(self.tmp_dir, file_name)        
        RACEParser.extract_article(
            os.path.join(self.input_dir, file_name),
            tmp_path
        )
                
        res = subprocess.run(
            [
                "/Users/YK/anaconda2/bin/python2",
                "parse.py",
                "-t",
                os.path.abspath(self.output_dir),
                tmp_path
            ],
            cwd="/Users/YK/mt/feng-hirst-rst-parser/src/"
        )
        return res.returncode
        
    def parse_texts(self):
        non_processed_files = []
        for f in os.listdir(self.input_dir):
            print(f"Parsing {f}..")
            if self.parse_text(f) != 0:
                non_processed_files.append(f)
        print("Done.")
        print("Unprocessed files: ")
        print("\n".join(non_processed_files))
    
#     def extract_articles(self, skip_files):
#         article_dir = os.path.join(self.tmp_dir, "articles")
#         os.makedirs(article_dir, exist_ok=True)
#         file_names = []
#         for f in os.listdir(self.input_dir):
#             if f not in skip_files:
#                 dest = os.path.join(article_dir, f)
#                 RACEParser.extract_article(
#                     os.path.join(self.input_dir, f),
#                     dest
#                 )
#                 file_names.append(os.path.abspath(dest))
#         return file_names
    
#     def parse_texts(self, skip_files):
#         tmp_path = os.path.join(self.tmp_dir, "file_names.txt")
        
#         with open(tmp_path, "wt") as f:
#             f.write(
#                 "\n".join(self.extract_articles(skip_files))
#             )
        
#         res = subprocess.run(
#             [
#                 "/Users/YK/anaconda2/bin/python2",
#                 "parse.py",
#                 "-tD",
#                 os.path.abspath(self.output_dir),
#                 tmp_path
#             ],
#             cwd="/Users/YK/mt/feng-hirst-rst-parser/src/",
#             stdin=subprocess.PIPE,
#             stderr=subprocess.PIPE
#         )
        
#         print("Done.")
```

```python
def get_unprocessed_files(subfolder):
    all_files = [
        f.split(".")[0] 
        for f in os.listdir(os.path.join("/Users/YK/mt/race/", subfolder))
    ]
    parsed_files = [
        f.split(".")[0] 
        for f in os.listdir(os.path.join("/Users/YK/mt/parsed/race/", subfolder))
    ]
    return set(all_files).difference(set(parsed_files))
```
